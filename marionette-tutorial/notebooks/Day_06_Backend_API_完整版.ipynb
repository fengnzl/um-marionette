## Part 1: FastAPI 基础

### 1.1 什么是 FastAPI？

**FastAPI** 是一个现代、快速的 Python Web 框架。

### 前端类比

| Express.js | FastAPI |
|------------|----------|
| app.get('/') | @app.get('/') |
| req.body | request: RequestModel |
| res.json() | return ResponseModel |
| async/await | async/await |

## Part 2: Pydantic 数据模型

### 2.1 定义请求和响应模型

````python
from pydantic import BaseModel, Field
from typing import List, Optional
from enum import IntEnum

# 定义枚举
class DayOfWeek(IntEnum):
    MONDAY = 25
    TUESDAY = 26
    # ... 其他天

# 定义请求模型
class GenerateRequest(BaseModel):
    condition1: int = Field(..., description="星期几")
    condition2: int = Field(32, ge=32, le=33, description="斋月状态")
    condition3: int = Field(34, ge=34, le=35, description="假期状态")
    num_samples: int = Field(10, ge=1, le=100, description="生成数量")
    tmax: float = Field(86400.0, description="最大时间")

# 定义响应模型
class GenerateResponse(BaseModel):
    success: bool
    message: str
    sequences: List[dict]
    generation_time: float

print("数据模型定义完成")
````

## Part 3: API 端点实现

### 3.1 核心 API 结构

````python
from fastapi import FastAPI, HTTPException
from fastapi.middleware.cors import CORSMiddleware
import uvicorn

# 创建应用
app = FastAPI(
    title="Marionette API",
    description="轨迹生成服务 API",
    version="1.0.0"
)

# 配置 CORS
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_methods=["*"],
    allow_headers=["*"],
)

# 全局变量
models_cache = {}

@app.get("/")
async def root():
    return {"message": "Marionette API", "version": "1.0.0"}

@app.get("/health")
async def health_check():
    return {"status": "healthy", "model_loaded": False}

print("FastAPI 应用结构定义完成")
````
