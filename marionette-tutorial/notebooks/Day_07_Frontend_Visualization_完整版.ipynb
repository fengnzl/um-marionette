## Part 1: Deck.gl 基础

### 1.1 什么是 Deck.gl？

**Deck.gl** 是 Uber 开发的 WebGL 地图可视化库。

### 与 ECharts 对比

| 特性 | ECharts | Deck.gl |
|------|---------|----------|
| 渲染引擎 | Canvas/SVG | WebGL |
| 数据规模 | 千级 | 百万级 |
| 地图支持 | 需要插件 | 原生支持 |

### 安装

```bash
npm install deck.gl @deck.gl/core @deck.gl/layers react-map-gl
```

## Part 2: 数据格式准备

### API 返回的数据格式

```json
{
  "success": true,
  "sequences": [
    {
      "sequence_id": "uuid-123",
      "points": [
        {
          "time": 3600,
          "poi_id": 10,
          "category": 0,
          "coordinate": {
            "lat": 41.0082,
            "lon": 28.9784
          }
        }
      ]
    }
  ]
}
```

### Deck.gl 需要的格式

**注意**: Deck.gl 使用 [lon, lat] 顺序，不是 [lat, lon]！

````python
# JavaScript 数据转换示例

conversion_example = '''
// 转换为 PathLayer 格式
const pathData = apiResponse.sequences.map(seq => ({
    path: seq.points
        .filter(p => p.coordinate)
        .map(p => [p.coordinate.lon, p.coordinate.lat])  // [lon, lat]
    color: [0, 128, 255, 200]
}));

// 转换为 ScatterplotLayer 格式
const scatterData = apiResponse.sequences.flatMap(seq =>
    seq.points
        .filter(p => p.coordinate)
        .map(p => ({
            coordinates: [p.coordinate.lon, p.coordinate.lat],
            color: [0, 128, 255, 200]
        }))
);
'''

print("JavaScript 数据转换示例:")
print(conversion_example)
````

## Part 3: 可视化组件

### 3.1 单屏地图组件

````python
react_component = '''
import DeckGL from '@deck.gl/react';
import { ScatterplotLayer, PathLayer } from '@deck.gl/layers';

function TrajectoryMap({ data }) {
    const [viewState, setViewState] = useState({
        longitude: 28.9784,
        latitude: 41.0082,
        zoom: 12
    });
    
    const layers = [
        new PathLayer({
            id: 'paths',
            data: preparePathData(data),
            getPath: d => d.path,
            getColor: [0, 128, 255, 200]
        }),
        new ScatterplotLayer({
            id: 'points',
            data: prepareScatterData(data),
            getPosition: d => d.coordinates,
            getFillColor: [0, 128, 255]
        })
    ];
    
    return (
        <DeckGL
            viewState={viewState}
            onViewStateChange={({ viewState }) => setViewState(viewState)}
            layers={layers}
            controller={true}
        />
    );
}
'''

print("React 组件示例:")
print(react_component)
````

## Part 5: 常见问题

### Q: CORS 错误
